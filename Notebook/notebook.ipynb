# `Solution : 1` Combine the directory data and url data get the resonse

```python
url= "https://www.brainwired.in/about-us"
from llama_index.readers.web import SimpleWebPageReader

doc2    = SimpleWebPageReader(html_to_text=True).load_data(urls=[url])


documents=SimpleDirectoryReader("../Data") # we are passing the pdf files
documents

doc = documents.load_data()
all_doc = doc2+doc

model=Gemini(models='gemini-pro',api_key=google_api_key)

gemini_embed_model=GeminiEmbedding(model_name="models/embedding-001")

# Configure Service Context
service_context = ServiceContext.from_defaults(llm=model,embed_model=gemini_embed_model, chunk_size=800, chunk_overlap=20)

index = VectorStoreIndex.from_documents(all_doc,service_context=service_context)


index.storage_context.persist()
query_engine=index.as_query_engine()
```

# Query engine


```python
query_engine=index.as_query_engine()
query_engine.query("GIVE THE BRAINWIRED FOUNDER NAME IN TABULAR FORMAT").response
```
